```python
from selenium import webdriver
from datetime import timedelta, datetime
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains 
```

```python
driver = webdriver.Chrome()
driver.get('https://www.coindesk.com/policy')
#html = driver.page_source
#soup = BeautifulSoup(html)
data = []

SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = 0
html = None
sleep_time = 1

while True:
    html = driver.page_source
    soup = BeautifulSoup(html)
    content = soup.find_all("div",attrs={"bg-white flex gap-6 w-full shrink justify-between"})
    time_data = soup.find_all("span",attrs={"class":"font-metadata text-color-charcoal-600 uppercase"}) 
    try:
        cur_year = int(time_data[-1].text[:4])
        if cur_year == 2019:
            break
    except:
        pass
    print(len(content))
    button = driver.find_elements(By.XPATH, '/html/body/div[1]/main/section/div/div[4]/div[2]/button')[0]
    bottom = driver.find_elements(By.XPATH, '/html/body/div[1]/main/footer')[0]
    #action = ActionChains(driver) 
    #action.move_to_element(button).click().perform()
    #driver.execute_script(f"window.scrollTo(0, {button.location['y']});")
    #button.click()
    action = ActionChains(driver) 
    action.move_to_element(bottom).perform()
    try:
        button.click()
        time.sleep(sleep_time)
    except:
        if sleep_time > 10:
           time.sleep(sleep_time) 
        else:
            sleep_time += 1
            print(sleep_time)
```
