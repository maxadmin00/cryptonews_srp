```python
from selenium import webdriver
from datetime import timedelta, datetime
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains 
```

```python
driver = webdriver.Chrome()
driver.get('https://www.cointelegraph.com/tags/investments')
#html = driver.page_source
#soup = BeautifulSoup(html)
data = []

SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = 0
html = None
sleep_time = 1

while True:
    html = driver.page_source
    soup = BeautifulSoup(html)
    content = soup.find_all("div",attrs={"class":"post-card-inline__header"})
    time_data = soup.find_all("div",attrs={"class":"post-card-inline__meta"}) 

    print(len(content))
    try:
        button = driver.find_elements(By.XPATH, '//*[@id="__layout"]/div/div[3]/div[4]/main/div/div/div[1]/div[3]/div/div/a')[0]
        bottom = driver.find_elements(By.XPATH, '//*[@id="__layout"]/div/div[3]/div[4]/main/div/div/div[1]/div[3]/div/div')[0]
        action = ActionChains(driver) 
        action.move_to_element(bottom).perform()
    except:
        bottom = driver.find_elements(By.CLASS_NAME, "post-card-inline__title")[-1]
        action = ActionChains(driver) 
        action.move_to_element(bottom).perform()
    #action = ActionChains(driver) 
    #action.move_to_element(button).click().perform()
    #driver.execute_script(f"window.scrollTo(0, {button.location['y']});")
    #button.click()
    try:
        button.click()
        time.sleep(sleep_time)
    except:
        time.sleep(sleep_time) 
```
